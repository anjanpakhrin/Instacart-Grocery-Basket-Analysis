#### **Step 4:** Create **loyalty flag** using *transform()* and *loc()*

#### Creating a flag for customers under following **criteria**:
- If **max_orders > 40**, then **"Loyal customer"**
- If **10 < max_orders <= 40**, then **"Regular customer"**
- If **max_orders <= 10**, then **"New customer"**

1. Split the data into groups based on "user_id" column.
2. Apply **transform()** function on the "order_number" column to generate the maximum orders for each user.
3. Create a new column, "max_order," to put the results of aggregation

#### Step 4.1: Deriving new column with **Transform()**

```python
# Create new column "max_order"
# Group by "user_id"
# Transform "order_number" by applying "max()" function from NumPy library
ords_prods_merge['max_order'] = ords_prods_merge.groupby('user_id')['order_number'].transform('max')
```

```python
# Checking output - displaying only 2 rows of each user limited to 10 rows in total.

ords_prods_merge.groupby('user_id').head(2).head(10)
```

#### Step 4.2: Deriving Columns with **loc()**
#### Creating new flagging column **loyalty_flag**

```python
# If max_order > 40, then "Loyal customer"
ords_prods_merge.loc[ords_prods_merge['max_order'] > 15, 'loyalty_flag'] = 'Loyal customer'

# If 10 < max_orders <= 40, then "Regular customer"
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order']> 10), 'loyalty_flag'] = 'Regular customer'

# If max_orders <= 10, then "New customer"
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 10, 'loyalty_flag')] = 'New customer'
```

```python
# Checking output for new column - displaying only 2 rows per category
ords_prods_merge.groupby('loyalty_flag').head(2)
```

```python
# Checking frequency for each category
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

**Majority of orders are placed by "Regular customer". Second highest orders are done by "Loyal customer". The lowest numbers of orders are placed by "New customer".** 

```python
# Counting the number of customers grouped by "loyalty_flag"

ords_prods_merge.groupby(['loyalty_flag']).user_id.nunique()
```

**"New customers" constitute Instacart's largest segment, followed by "Loyal customers" as the second-largest group, and "Regular customers" as the third-largest.**

#### **Step 5:** Determining purchasing price difference by customer category - **loyalty_flag**

```python
# Step 1: Calculating key summary statistics grouped by "loyalty_flag"
loyalty_price_stats = ords_prods_merge.groupby('loyalty_flag')['prices'].agg(['mean', 'median', 'min', 'max', 'std'])

# Return statistics
loyalty_price_stats
```

#### Analysis of purchasing behavior based on prices for different customer categories
* Based on above statistics, **loyal customers spend less** compared to new and regular customer.
* New customers spend the most.
* Same median values (7.4) suggest that the majority of purchases is very similar, regardless of loyalty status.
* Significant difference in **min** and **max** for all groups shows that the dataset contains some **extrem outliers**, which might be data entry errors or rare products prices. This obviously affects the mean and standard deviation.
* Standard deviation is much higher for new and regular customers, This means New and regular customers' spending behavior is less predictable. They might be trying a wide variety of products at very different price points. In opposite, loyal customers have likely settled into a routine of buying thier regular, known products.

#### **Step 6:** Identifying different type of spenders under following criteria:
* **Low spender:** Users with mean product price < 10.
  * **Condition:** user_mean_prices < 10
* **High spender:** Users wuth mean product price is greater or equal to 10.
  * **Condition:** user_mean_prices >= 10

#### Step 6.1: Aggregating with transform() --> Creating **"user mean prices"**
* Split data into groups based on **"user_id"**.
* Apply **transform()** on **"prices"** column to generate the mean of the prices of products purchased.
* Create a new column **"user_mean_prices"** to put the results

```python
# Create ne column "user_mean_prices"
# Group by "user_id"
# Transform "prices" by "mean()" function

ords_prods_merge['user_mean_prices'] = ords_prods_merge.groupby('user_id')['prices'].transform('mean')
```

```python
# Checking output

ords_prods_merge.groupby('user_id').head(2).head(10)
```

#### **Summary to above table**
* Appended a new column **"user_mean_prices"** based on mean user item price.
* Calculated **mean price per user** and assigned the value to all user rows.
* Diplayed 2 rows per user for the first 5 users, limited to a total 10 rows for a **preliminary review** of the new column.

#### Step 6.2: Deriving column with loc() --> Creating **"Spending flag"**
#### Criteria:
* If **user_mean_prices** < 10, then **"Low spender"**
* if **user_mean_prices** >= 10, then **"High spender"**

```python
# If user_mean_prices < 10, then "Low spender"
ords_prods_merge.loc[ords_prods_merge['user_mean_prices'] < 10, 'spending_flag'] = 'Low spender'

# If user_mean_prices >= 10, then "High spender"
ords_prods_merge.loc[ords_prods_merge['user_mean_prices'] >= 10, 'spending_flag'] = 'High spender'
```

#### **Step 7:** Identifying **frequent**, **non-frequent**, and **regular** customers
#### Criteria:
* If median of **days_since_prior_order > 20**, then **"Non-frequent customer"**
* If median of **10 < days_since_prior_order <= 20**, then **"Regular customer"**
* If median of **days_since_prior_order <= 10**, then **"Frequent customer"**

#### Step 7.1: Calculation **median** for **days_since_prior_order**

```python
# Create new column "median_frequency" for median for days_since_prior_order
# Group by "user_id"
# Transform "days_since_prior_order" by applying "max()" function
ords_prods_merge['median_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
# Checking output

ords_prods_merge.groupby('user_id').head(2).head(10)
```

#### Step 7.2: Creating flagging column **frequency_flag**

```python
# If median_frequency > 20, then "Non-frequent customer"
ords_prods_merge.loc[ords_prods_merge['median_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'

# If 10 < median_days_since_prior_order <= 20, then "Regular customer"
ords_prods_merge.loc[(ords_prods_merge['median_frequency'] <= 20) & (ords_prods_merge['median_frequency']> 10), 'frequency_flag'] = 'Regular customer'

# If max_orders <= 10, then "New customer"
ords_prods_merge.loc[(ords_prods_merge['median_frequency'] <= 10, 'frequency_flag')] = 'Frequent customer'
```

```python
# Checking output for new column
ords_prods_merge.groupby('frequency_flag').head(2)
```
