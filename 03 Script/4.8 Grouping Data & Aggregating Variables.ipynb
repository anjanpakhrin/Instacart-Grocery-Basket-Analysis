## **Aggregating Data with agg()**

#### - Performing a Single Aggregation

```python
# Grouping by "department_id" and calculating mean values for "order_number"

df.groupby('department_id').agg({'order_number': ['mean']})
```

**This shows the average number of orders per user for each department ID.**

```python
# Calculating "aagregate" without using "agg()"

df.groupby('department_id')['order_number'].mean()
```

#### **DOT NOTATION VS. SQUARE BRACKETS**
**DOT NOTATION:**
- **Main job:** Doing actions & asking for info (a multi-tool)
- **Best for:** Using fuctions *df.means()* and attributes *df.shape*

**SQUARE BRACKETS:**
- **Main job:** Selecting data (a specialist)
- **Best for:** Picking columns *df['column']* and rows *df[0:5]*
- Clearer and faster

```python
# Calculate "mean" with "DOT NOTATION"

df.groupby('department_id').order_number.mean()
```

```python
# Calculate "mean" with "SQUARE BRACKETS"

df.groupby('department_id')['order_number'].mean()
```

#### - Performing a Multiple Aggregations

```python
# Calculation "mean", "min", and "max" of "order_number"

df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

## **Aggregating Data with transform()**

#### Creating a flag for customers under following **criteria**:
- If **max_orders > 40**, then **"Loyal customer"**
- If **10 < max_orders <= 40**, then **"Regular customer"**
- If **max_orders <= 10**, then **"New customer"**

1. Split the data into groups based on "user_id" column.
2. Apply **transform()** function on the "order_number" column to generate the maximum orders for each user.
3. Create a new column, "max_order," to put the results of aggregation

```python
# Create new column "max_order"
# Group by "user_id"
# Transform "order_number" by applying "max()" function from NumPy library
ords_prods_merge['max_order'] = ords_prods_merge.groupby('user_id')['order_number'].transform('max')
```

```python
# Checking output

ords_prods_merge.head(10)
```

## Deriving Columns with loc()

#### Creating new flagging column in dataframe "ords_prods_merge"

```python
# If max_order > 40, then "Loyal customer"
ords_prods_merge.loc[ords_prods_merge['max_order'] > 15, 'loyalty_flag'] = 'Loyal customer'

# If 10 < max_orders <= 40, then "Regular customer"
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order']> 10), 'loyalty_flag'] = 'Regular customer'

# If max_orders <= 10, then "New customer"
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 10, 'loyalty_flag')] = 'New customer'
```

#### Identifying **OUTLIERS**
* A value of 100 for prices can be defined as threshold assuming that an item in a supermarket would cost in average $100.
* Anything above this threshold could be considered as an outlier.

```python
# Running the threshold check - Outlier observation

ords_prods_merge.loc[ords_prods_merge['prices'] > 100]
```

#### **Observations:**
* There are 5,127 rows with outlier observations (prices > $100).
* Prices like $14,900.00 or $99,9999 per item don't make sense in terms of the other values in column.
* Values > 100, therefore, defined as outliers and marked as missing (NaN).
#### Turning values > 100 into NaNs

```python
# Turning outliers into NaNs

ords_prods_merge.loc[ords_prods_merge['prices'] > 100, 'prices'] = np.nan
```
